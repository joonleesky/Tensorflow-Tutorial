```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("../data/mnist/", one_hot = True)
```

### Build Graph

```python
# Config
learning_rate  = 2e-4
training_epoch = 100
batch_size = 128

hidden_size = 256
input_size  = 28*28
noise_size  = 128
class_size  = 10

sample_size = 10
```

```python
X = tf.placeholder(tf.float32, [None, input_size])
Y = tf.placeholder(tf.float32, [None, class_size])
Z = tf.placeholder(tf.float32, [None, noise_size])
```

```python
def get_random_noise(batch_size, noise_size):
    
    return np.random.normal(size = (batch_size, noise_size))
```

```python
def generator(noise, labels):
    with tf.variable_scope('generator'):
        inputs = tf.concat([noise,labels],1)
        # tf.layers.dense = tf.variable Weight/Bias + tf.matmul
        hidden = tf.layers.dense(inputs, hidden_size, activation = tf.nn.relu)
        output = tf.layers.dense(hidden, input_size,  activation = tf.nn.sigmoid)

    return output
```

```python
def discriminator(inputs, labels, reuse = None):
    with tf.variable_scope('discriminator') as scope:
        if reuse:
            scope.reuse_variables()
        inputs = tf.concat([inputs,labels],1)
        hidden = tf.layers.dense(inputs, hidden_size, activation = tf.nn.relu)
        output = tf.layers.dense(hidden, 1,  activation = None)

    return output
```

```python
G = generator(Z, Y)
D_real = discriminator(X, Y)
D_fake = discriminator(G, Y, True) 
# If a variable scope, reuse_variables() is not declared
#  D_real and D_fake will get different variable sets

print(G)
print(D_real)
print(D_fake)
```

```python
# Discriminator: true image should be closed to 1, fake image should be closed to 0
loss_D_real = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits = D_real, labels = tf.ones_like(D_real)))
loss_D_fake = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits = D_fake, labels = tf.zeros_like(D_fake)))

loss_D = loss_D_real + loss_D_fake
```

```python
# Generator: fake image should be closed to 1
loss_G = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits = D_fake, labels = tf.ones_like(D_fake)))
```

```python
# Designate which variables to train in each loss function
variables_D = tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES, scope = 'discriminator')
variables_G = tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES, scope = 'generator')

train_D = tf.train.AdamOptimizer(learning_rate).minimize(loss_D, var_list = variables_D)
train_G = tf.train.AdamOptimizer(learning_rate).minimize(loss_G, var_list = variables_G)
```

### Train Graph

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    total_batch = int(mnist.train.num_examples / batch_size)
    
    for epoch in range(training_epoch):
        total_loss_D, total_loss_G = 0, 0
        
        for _ in range(total_batch):
            batch_xs, batch_ys = mnist.train.next_batch(batch_size)
            noise = get_random_noise(batch_size, noise_size)
            _, loss_val_D = sess.run([train_D, loss_D], feed_dict = {X:batch_xs, Y:batch_ys, Z:noise})
            _, loss_val_G = sess.run([train_G, loss_G], feed_dict = {Y:batch_ys, Z:noise})
            
            total_loss_D += loss_val_D
            total_loss_G += loss_val_G
    
        print('Epoch:','%04d' %(epoch+1),
              'D Loss: %.3f' %(total_loss_D / total_batch),
              'G Loss: %.3f' %(total_loss_G / total_batch))
        
        if epoch == 0 or (epoch + 1) % 5 == 0:
            sample_size = 10
            noise = get_random_noise(sample_size, noise_size)
            
            samples = sess.run(G, feed_dict={Y:mnist.test.labels[:sample_size], Z:noise})
            
            fig, ax = plt.subplots(2, sample_size, figsize = (sample_size, 2))
            
            for i in range(sample_size):
                ax[0][i].set_axis_off()
                ax[1][i].set_axis_off()
                
                ax[0][i].imshow(np.reshape(mnist.test.images[i],(28,28)))
                ax[1][i].imshow(np.reshape(samples[i],(28,28)))
            plt.show()
```
