```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("../data/mnist/", one_hot = True)
```

```python
# Config
learning_rate  = 2e-4
training_epoch = 100
batch_size = 128

hidden_size = 256
input_size  = 28*28
noise_size  = 128

sample_size = 10
```

### Build Graph

```python
X = tf.placeholder(tf.float32, [None, input_size])
Z = tf.placeholder(tf.float32, [None, noise_size])
```

```python
# Generator
G_W1 = tf.Variable(tf.random_normal([noise_size, hidden_size],stddev = 0.01))
G_b1 = tf.Variable(tf.zeros([hidden_size]))

G_W2 = tf.Variable(tf.random_normal([hidden_size, input_size],stddev = 0.01))
G_b2 = tf.Variable(tf.zeros([input_size]))

G_var_list = [G_W1,G_b1,G_W2,G_b2]
```

```python
# Discriminator
D_W1 = tf.Variable(tf.random_normal([input_size, hidden_size],stddev = 0.01))
D_b1 = tf.Variable(tf.zeros([hidden_size]))

D_W2 = tf.Variable(tf.random_normal([hidden_size, 1],stddev = 0.01))
D_b2 = tf.Variable(tf.zeros([1]))

D_var_list = [D_W1,D_b1,D_W2,D_b2]
```

```python
def generator(noise):
    hidden = tf.nn.relu(tf.matmul(noise, G_W1) + G_b1)
    output = tf.nn.sigmoid(tf.matmul(hidden, G_W2) + G_b2)

    return output
```

```python
def discriminator(image):
    hidden = tf.nn.relu(tf.matmul(image, D_W1) + D_b1)
    output = tf.nn.sigmoid(tf.matmul(hidden, D_W2) + D_b2)

    return output
```

```python
def get_random_noise(batch_size, noise_size):
    
    return np.random.normal(size = (batch_size, noise_size))
```

```python
G = generator(Z)
D_fake = discriminator(G)
D_real = discriminator(X)
```

```python
D_loss = -tf.reduce_mean(tf.log(D_real) + tf.log(1 - D_fake)) #D_real should become 1 and D_fake should become 0
G_loss = -tf.reduce_mean(tf.log(D_fake)) # Trying to make D_fake as 1
```

```python
train_G = tf.train.AdamOptimizer(learning_rate).minimize(G_loss, var_list = G_var_list)
train_D = tf.train.AdamOptimizer(learning_rate).minimize(D_loss, var_list = D_var_list)
```

### Run Graph

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    total_batch = int(mnist.train.num_examples / batch_size)
    
    for epoch in range(training_epoch):
        total_loss_D, total_loss_G = 0, 0
        
        for _ in range(total_batch):
            batch_xs, batch_ys = mnist.train.next_batch(batch_size)
            noise = get_random_noise(batch_size, noise_size)
            
            _, loss_val_D = sess.run([train_D, D_loss], feed_dict = {X:batch_xs, Z:noise})
            _, loss_val_G = sess.run([train_G, G_loss], feed_dict = {Z:noise})
            
            total_loss_D += loss_val_D
            total_loss_G += loss_val_G
    
        print('Epoch:','%04d' %(epoch+1),
              'D Loss: %.3f' %(total_loss_D / total_batch),
              'G Loss: %.3f' %(total_loss_G / total_batch))
        
        if epoch == 0 or (epoch + 1) % 5 == 0:
            sample_size = 10
            noise = get_random_noise(sample_size, noise_size)
            
            samples = sess.run(G, feed_dict={Z:noise})
            
            fig, ax = plt.subplots(1, sample_size, figsize = (sample_size, 1))
            
            for i in range(sample_size):
                ax[i].set_axis_off()
                ax[i].imshow(np.reshape(samples[i],(28,28)))
            plt.show()
```
